```python

"""
Sistema para encontrar una ruta entre un punto A(origen) y un punto B(destino)
en un sistema de transporte masivo.

Se usa un algoritmo de búsqueda simple.
"""

# =========================================================
# 1. DATASET - Datos no reales
# Estación/conexiones entre estaciones/ tiempo estimado entre la estación origen y destinos
# =========================================================
red_transporte = {
    "Portal Norte": [("Calle 100", 10), ("Usaquén", 15)],
    "Calle 100": [("Portal Norte", 10), ("Héroes", 8)],
    "Héroes": [("Calle 100", 8), ("Calle 72", 7)],
    "Calle 72": [("Héroes", 7), ("Calle 45", 9), ("Chapinero", 6)],
    "Calle 45": [("Calle 72", 9), ("Av. Jiménez", 10)],
    "Av. Jiménez": [("Calle 45", 10), ("Centro", 5)],
    "Centro": [("Av. Jiménez", 5), ("Museo Nacional", 7)],
    "Museo Nacional": [("Centro", 7)],
    "Chapinero": [("Calle 72", 6)],
    "Usaquén": [("Portal Norte", 15)]
}

# =========================================================
# 2. NUMERACIÓN DE ESTACIONES
# =========================================================
estaciones = list(red_transporte.keys())
mapa_numero_estacion = {i + 1: estacion for i, estacion in enumerate(estaciones)}

# =========================================================
# 3. MOSTRAR ESTACIONES
# =========================================================
def mostrar_estaciones():
    print("\nESTACIONES DISPONIBLES:")
    for numero, estacion in mapa_numero_estacion.items():
        print(f"{numero}. {estacion}")

# =========================================================
# 4. BÚSQUEDA SIMPLE
# =========================================================
def busqueda_simple(red, origen, destino):
    """
    Búsqueda simple,recorre la red desde el origen hasta encontrar el destino.

    Retorna:
    - ruta encontrada
    - tiempo total de esa ruta
    """
    cola = [(origen, [origen], 0)]
    visitados = []

    paso = 1

    print("\n" + "=" * 70)
    print("EXPLICACIÓN DEL PROCESO DE BÚSQUEDA SIMPLE")
    print("=" * 70)
    print(f"Origen seleccionado: {origen}")
    print(f"Destino seleccionado: {destino}")
    print("\nLa búsqueda simple revisa estaciones vecinas una a una")
    print("hasta encontrar una ruta disponible hacia el destino.\n")

    while cola:
        estacion_actual, camino, tiempo_acumulado = cola.pop(0)

        print("-" * 70)
        print(f"PASO {paso}")
        print(f"Se toma de la cola la estación: {estacion_actual}")
        print(f"Camino actual: {' -> '.join(camino)}")
        print(f"Tiempo acumulado: {tiempo_acumulado} minutos")

        if estacion_actual == destino:
            print("\n" + "=" * 70)
            print("DECISIÓN FINAL DEL MODELO")
            print("=" * 70)
            print("Se encontró el destino.")
            print("La búsqueda simple devuelve el primer camino válido encontrado.")
            print(f"Ruta encontrada: {' -> '.join(camino)}")
            print(f"Tiempo total de la ruta: {tiempo_acumulado} minutos")
            return camino, tiempo_acumulado

        if estacion_actual not in visitados:
            visitados.append(estacion_actual)
            print(f"Se marca {estacion_actual} como visitada.")

            print("\nSe revisan las estaciones vecinas:")
            for vecino, tiempo in red.get(estacion_actual, []):
                if vecino not in visitados and vecino not in camino:
                    nuevo_camino = camino + [vecino]
                    nuevo_tiempo = tiempo_acumulado + tiempo

                    print(f"  - Vecino encontrado: {vecino}")
                    print(f"    Tiempo del tramo {estacion_actual} -> {vecino}: {tiempo} min")
                    print(f"    Nuevo camino propuesto: {' -> '.join(nuevo_camino)}")
                    print(f"    Tiempo acumulado del nuevo camino: {nuevo_tiempo} min")

                    cola.append((vecino, nuevo_camino, nuevo_tiempo))
                    print("    Se agrega este camino a la cola.")

        print("\nEstado actual de la cola:")
        if cola:
            for item in cola:
                print(f"  Estación: {item[0]} | Camino: {' -> '.join(item[1])} | Tiempo: {item[2]} min")
        else:
            print("  La cola está vacía.")

        paso += 1

    print("\nNo se encontró una ruta disponible.")
    return None, None

# =========================================================
# 5. PROGRAMA PRINCIPAL
# =========================================================
print("=" * 70)
print("SISTEMA INTELIGENTE DE RUTAS EN TRANSPORTE MASIVO")
print("=" * 70)
print("Algoritmo de búsqueda simple.\n")

mostrar_estaciones()

try:
    origen_num = int(input("\nIngrese el número de la estación de origen: "))
    destino_num = int(input("Ingrese el número de la estación de destino: "))

    if origen_num not in mapa_numero_estacion or destino_num not in mapa_numero_estacion:
        print("\nError: uno de los números ingresados no corresponde a una estación válida.")
    else:
        origen = mapa_numero_estacion[origen_num]
        destino = mapa_numero_estacion[destino_num]

        ruta, tiempo = busqueda_simple(red_transporte, origen, destino)

        print("\n" + "=" * 70)
        print("RESULTADO FINAL")
        print("=" * 70)

        if ruta:
            print(f"Origen: {origen}")
            print(f"Destino: {destino}")
            print(f"Ruta encontrada: {' -> '.join(ruta)}")
            print(f"Tiempo total estimado: {tiempo} minutos")

        else:
            print("No se encontró una ruta entre las estaciones seleccionadas.")

except ValueError:
    print("\nError: debes ingresar únicamente números enteros válidos.")
```
